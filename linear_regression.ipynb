```python
import sklearn
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
cars = pd.read_csv('mtcars.csv')
```

```python
cars.head()
```

```python
cars.shape
```

```python
#Linear Regression to predict mileage mpg using other variable

#mpg is the target variable
```

```python
# checking for assumptions of linear Reg 1. Linearity of predictor variables with the target

cars.corr()
```

```python
# checking for assumptions of linear Reg 2. Normality, we shall normalize univariate, multivariate is assumed with higher dimession

%matplotlib inline
sb.distplot(cars.hp)

print('range of Skewness, S < 1.96')
print('Skewness of hp:', stats.skew(cars.hp))
print('Skewness of wt:', stats.skew(cars.wt))
print('Kurtosis of hp:', stats.kurtosis(cars.hp))
print('Kurtosis of wt:', stats.kurtosis(cars.wt))
```

```python
X = cars.loc[ : , ['hp','wt','am']]
y = cars.mpg
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.15, random_state = 14)
```

```python
print('X_train Shape', X_train.shape)
print('X_test Shape', X_test.shape)
print('y_train Shape', y_train.shape)
print('y_test Shape', y_test.shape)
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(X_train,y_train)
```

```python
y_predict = model_lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_predict)
```
